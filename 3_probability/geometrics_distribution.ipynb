# Geometric Distribution: Introduction

The **Geometric Distribution** models the number of trials required to achieve the first success in a sequence of independent trials, each with the same probability of success.

For example, it can be used to model how many attempts are needed to get a heads in a coin flip or how many customers need to be contacted before making the first sale.

--- 

```python
# Import necessary libraries
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Parameters for the geometric distribution
p_success = 0.3  # Probability of success in each trial
n_experiments = 1000  # Number of experiments

# Generate geometric distribution data
geometric_data = np.random.geometric(p=p_success, size=n_experiments)
```

```python
# Plot the Geometric Distribution
plt.figure(figsize=(10, 6))

# Setting the bins to correspond to integer values
bins = np.arange(geometric_data.min(), geometric_data.max() + 1) - 0.5

# Plot the histogram with raw counts
sns.histplot(geometric_data, bins=bins, kde=False, stat='count', color='skyblue')

plt.title(f"Geometric Distribution (p={p_success}) - 1000 Experiments")
plt.xlabel("Number of Trials Until First Success")
plt.ylabel("Frequency (Number of Experiments)")
plt.grid(True)
plt.show()
```

```python
# Check total number of experiments to ensure correctness
print(f"Total number of experiments: {n_experiments}")
print(f"Sum of frequencies: {len(geometric_data)}")
```

# Interpretation

The plot shows the number of trials required to achieve the first success. With a probability of success of `p=0.3`, most trials result in a success within the first few attempts. As expected, the geometric distribution is right-skewed, meaning there are fewer trials requiring a large number of attempts to reach success.

In practice, the geometric distribution is useful in scenarios where we want to model the likelihood of waiting for an event to occur, such as the first customer response in a marketing campaign or the first defective product in a manufacturing process.
