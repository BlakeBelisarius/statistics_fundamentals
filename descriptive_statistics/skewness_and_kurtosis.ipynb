##### In this notebook, we'll be calculating and visualizing skewness and kurtosis of a dataset using scipy.stats.skew() and scipy.stats.kurtosis(). These measures are essential for determining the symmetry and tail behavior of a distribution, which can affect both descriptive and inferential statistics.
---

Steps:
 + Generate a normal and a skewed dataset.
 + Calculate skewness and kurtosis.
 + Visualize the distributions and show the calculated skewness and kurtosis values.

```python
# Importing Libraries

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis
```

```python
# Generate a normally distributed dataset
np.random.seed(0)
normal_data = np.random.normal(loc=50, scale=10, size=1000)

# Generate a right-skewed dataset
skewed_data = np.random.exponential(scale=2, size=1000)
```

```python
# Calculate skewness and kurtosis for both datasets
normal_skewness = skew(normal_data)
normal_kurtosis = kurtosis(normal_data)

skewed_skewness = skew(skewed_data)
skewed_kurtosis = kurtosis(skewed_data)
```

```python
# Print skewness and kurtosis values
print(f"Normal Data Skewness: {normal_skewness:.2f}, Kurtosis: {normal_kurtosis:.2f}")
print(f"Skewed Data Skewness: {skewed_skewness:.2f}, Kurtosis: {skewed_kurtosis:.2f}")
```

```python
# Plot normal distribution
plt.subplot(1, 2, 1)
sns.histplot(normal_data, kde=True, color='blue')
plt.title(f'Normal Data\nSkewness: {normal_skewness:.2f}, Kurtosis: {normal_kurtosis:.2f}')

# Plot skewed distribution
plt.subplot(1, 2, 2)
sns.histplot(skewed_data, kde=True, color='green')
plt.title(f'Skewed Data\nSkewness: {skewed_skewness:.2f}, Kurtosis: {skewed_kurtosis:.2f}')

plt.tight_layout()
plt.show()
```

Explanation:
1. Data Generation: 
   + Normal Data: A dataset generated from a normal distribution.
   + Skewed Data: A right-skewed dataset generated using an exponential distribution.

2. Skewness & Kurtosis:
   + Skewness: Measures the asymmetry of the distribution. A skewness of 0 indicates perfect symmetry.
       + Positive skewness means a longer right tail (right-skewed).
      + Negative skewness means a longer left tail (left-skewed).
   + Kurtosis: Measures the "tailedness" of the distribution. Higher kurtosis indicates more data in the tails.
       + Kurtosis of 3 (excess kurtosis = 0) indicates a normal distribution.
      + Positive kurtosis means heavier tails (leptokurtic).
      + Negative kurtosis means lighter tails (platykurtic).
3. Visualization:
    + Two subplots showing the distributions of both datasets, with skewness and kurtosis values overlaid in the titles.

Why This Matters:
  + Skewness: Indicates if the data is symmetrically distributed or if it has a longer tail on one side, which can skew the results of statistical tests.  
  + Kurtosis: Helps identify if the data has outliers or is more concentrated in the tails, which can affect the reliability of statistical measures like the standard deviation.
