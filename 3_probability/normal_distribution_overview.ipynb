# Normal Distribution: An Overview
The **normal distribution** is a symmetrical, bell-shaped curve representing many natural phenomena, such as heights, test scores, and IQs. In a normal distribution:
- Most of the data points cluster around the mean.
- The distribution is symmetrical about the mean, with the left and right sides mirroring each other.

In this notebook, we will:
1. Generate a normal distribution using Python.
2. Visualize the distribution with a histogram and a density curve.
3. Calculate probabilities associated with the normal distribution using the CDF (Cumulative Distribution Function).

---

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

## Generating a Normal Distribution
We will use NumPy to generate a set of data points that follow a normal distribution. We will generate 1000 random values with:
- **Mean**: 0
- **Standard Deviation**: 1


```python
# Generating 1000 random data points from a normal distribution
mean = 0
std_dev = 1
n_samples = 1000

# Using NumPy's random.normal() function to generate normal distribution data
data = np.random.normal(mean, std_dev, n_samples)
```

## Visualizing the Normal Distribution
To better understand the data, we can visualize it using a **histogram** and overlay a **probability density function (PDF)** to illustrate the shape of the normal distribution.


```python
# Plot the data to visualize the normal distribution
plt.figure(figsize=(10, 6))
sns.histplot(data, bins=30, kde=True, stat="density", linewidth=0)

# Overlay the theoretical normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2)

# Adding titles and labels
plt.title('Normal Distribution: Mean = 0, Std Dev = 1')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

## Calculating Probabilities using the CDF
The **Cumulative Distribution Function (CDF)** gives the probability that a random variable drawn from the normal distribution will be less than or equal to a given value.

In this section, we'll calculate:
1. The probability that a value is less than 1 (\(P(X < 1)\)).
2. The probability that a value is between -1 and 1 (\(P(-1 < X < 1)\)).

```python
# Calculate probability that a value is less than 1 (P(X < 1))
probability_less_than_1 = norm.cdf(1, mean, std_dev)
print(f"Probability of a value less than 1 in the normal distribution: {probability_less_than_1:.4f}")

# Calculate probability that a value is between -1 and 1 (P(-1 < X < 1))
probability_between_neg1_and_1 = norm.cdf(1, mean, std_dev) - norm.cdf(-1, mean, std_dev)
print(f"Probability of a value between -1 and 1: {probability_between_neg1_and_1:.4f}")
```
