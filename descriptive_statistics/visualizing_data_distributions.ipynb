Descriptive statistics often assume that data follows a certain distribution, typically a normal distribution. Always visualize your data using histograms or kernel density plots (seaborn.distplot()) to verify the distribution. For non-normal data, consider transformations (e.g., logarithmic or square root) to normalize the distribution.   

In this notebook, we'll be visualizing the distribution of data using histograms and Kernel Density Estimation (KDE) plots. Additionally, we'll apply logarithmic and square root transformations to show how they can help normalize non-normal data distributions.

---

Steps:
 + Generate a random dataset (both normally distributed and skewed).
 + Create visualizations for:
   + A histogram and KDE plot.
   + Logarithmic and square root transformations to adjust skewed data

```python
# Import Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# Generate a random dataset (normally distributed)
np.random.seed(0)
normal_data = np.random.normal(loc=50, scale=10, size=1000)

# Generate a skewed dataset (right skewed)
skewed_data = np.random.exponential(scale=2, size=1000)
```

```python
# Plot original normal distribution
plt.figure(figsize=(14, 8))

plt.subplot(2, 2, 1)
sns.histplot(normal_data, kde=True, color='blue')
plt.title('Normal Distribution: Histogram & KDE')

# Plot original skewed data
plt.subplot(2, 2, 2)
sns.histplot(skewed_data, kde=True, color='green')
plt.title('Skewed Data: Histogram & KDE')

# Apply log transformation
log_data = np.log(skewed_data + 1)  # Adding 1 to avoid log(0)

plt.subplot(2, 2, 3)
sns.histplot(log_data, kde=True, color='red')
plt.title('Log Transformed Data: Histogram & KDE')

# Apply square root transformation
sqrt_data = np.sqrt(skewed_data)

plt.subplot(2, 2, 4)
sns.histplot(sqrt_data, kde=True, color='purple')
plt.title('Square Root Transformed Data: Histogram & KDE')

# Display the plots
plt.tight_layout()
plt.show()
```

Explanation:
 1. Histograms & KDE: Visualize the original distributions for both normal and skewed datasets.
    + Normal data is generated using np.random.normal(), which typically follows a bell curve.
    + Skewed data is generated using np.random.exponential(), which produces right-skewed data.

2. Transformations:
    + Logarithmic Transformation: Applied using np.log(). This compresses the data, reducing skewness.
    + Square Root Transformation: Applied using np.sqrt(). This is another common transformation used to reduce right-skewed distributions.

Why These Visualizations Matter:  
+ Histogram & KDE: Help to quickly check whether the data follows a normal distribution or is skewed.  
+ Transformations: For non-normal data, log or square root transformations can help normalize the distribution, which is critical for many statistical analyses that assume normality.
