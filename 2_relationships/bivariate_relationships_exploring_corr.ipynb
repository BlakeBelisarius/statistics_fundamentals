# Bivariate Relationships: Exploring Correlation

In statistics, **bivariate relationships** describe the association between two variables. A common measure used to quantify this relationship is the **Pearson correlation coefficient**. This coefficient measures the strength and direction of a linear relationship between two continuous variables.

In this notebook, we will:
1. Visualize the relationship between two variables using a **scatter plot**.
2. Calculate the **Pearson correlation coefficient**.
3. Interpret the results and discuss how relationships between variables can be used in analysis.


```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Enable plots in the notebook
%matplotlib inline
```

## Loading a Sample Dataset

For this demonstration, we will use the **Iris dataset**, which contains measurements of sepal length, sepal width, petal length, and petal width of various Iris flower species. We will examine the relationship between **sepal length** and **petal length**.


```python
# Load the built-in Iris dataset from seaborn
iris = sns.load_dataset("iris")

# Preview the dataset
iris.head()
```

## Visualizing the Relationship with a Scatter Plot

A **scatter plot** allows us to visualize the relationship between two continuous variables. In this case, we will plot **sepal length** against **petal length** to see if there is a noticeable relationship.

```python
# Create a scatter plot to visualize the relationship between sepal length and petal length
plt.figure(figsize=(8, 6))
sns.scatterplot(x="sepal_length", y="petal_length", data=iris)

# Add titles and labels
plt.title('Scatter Plot: Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.grid(True)
plt.show()
```

## Pearson Correlation Coefficient

The **Pearson correlation coefficient** quantifies the strength of the linear relationship between two variables. The coefficient ranges from -1 to 1:
- **1**: Perfect positive linear relationship.
- **-1**: Perfect negative linear relationship.
- **0**: No linear relationship.

Let's calculate the Pearson correlation coefficient between **sepal length** and **petal length**.


```python
# Calculate the Pearson correlation coefficient and p-value
corr_coef, p_value = pearsonr(iris["sepal_length"], iris["petal_length"])

# Display the results
print(f"Pearson Correlation Coefficient: {corr_coef:.4f}")
print(f"P-value: {p_value:.4e}")
```

## Interpretation

The **Pearson correlation coefficient** between **sepal length** and **petal length** is a positive value, suggesting a positive linear relationship between these two variables. In this case, as the sepal length increases, the petal length tends to increase as well.

The **p-value** is very small, indicating that the relationship is statistically significant.

### Key Points:
- A positive correlation implies that the two variables tend to increase together.
- A correlation does not imply **causation**, but it can reveal strong linear relationships.
